```python

## ----- CASE FOLDING ----- ##
```

```python

# 1. Mengubah text menjadi lowercase
kalimat = "5.0 The egg martabak is tasty, you can request a tipker for the Bandung cake, aka dry thin (Original) Martabak telornya gurih, bisa request tipker buat kue bandungnya alias tipis kering"
lower_case = kalimat.lower()
print(lower_case)
```

```python
# 2. Menghapus angka
import re #impor modul regular expressio
kalimat = "5.0 The egg martabak is tasty, you can request a tipker for the Bandung cake, aka dry thin (Original) Martabak telornya gurih, bisa request tipker buat kue bandungnya alias tipis kering"
hasil = re.sub(r"\d+","", kalimat)
print (hasil)
```

```python
# 3. Menghapus tanda baca
import string
kalimat = "5.0 The egg martabak is tasty, you can request a tipker for the Bandung cake, aka dry thin (Original) Martabak telornya gurih, bisa request tipker buat kue bandungnya alias tipis kering"
hasil = kalimat.translate(str.maketrans("","",string.punctuation))
print (hasil)
```

```python
# impor word_tokenize dari modul nltk
from nltk.tokenize import sent_tokenize, word_tokenize
kalimat = "5.0 The egg martabak is tasty, you can request a tipker for the Bandung cake, aka dry thin (Original) Martabak telornya gurih, bisa request tipker buat kue bandungnya alias tipis kering"
tokens = word_tokenize(kalimat)
print(tokens)
```

```python
# Daftar stopword yang tersedia di library Sastrawi
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
factory = StopWordRemoverFactory()
stopwords = factory.get_stop_words()
print(stopwords)
```

```python

# imort StopWordRemoverFactory class
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()
# Kalimat
kalimat = "Martabak telornya gurih, bisa request tipker buat kue bandungnya alias tipis kering"
stop = stopword.remove(kalimat)
print(stop)
```
